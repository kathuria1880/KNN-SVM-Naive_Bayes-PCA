```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('wine.csv')
```

```python
data.head()
```

```python
X=data.drop('Wine',1)
y=data['Wine']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
######################################  NORMALISING DATA  ############################################################## 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
from sklearn.decomposition import PCA
pca=PCA()
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
############################################   APPLYING PCA   #############################################################
explained_variance = pca.explained_variance_ratio_
for i in explained_variance:
    print(i)
evr=[]
for i in explained_variance:
    evr.append(float(round(i*100,3)))
######################################### EXPLAINED VARIANCES VS. COLUMNS  ################################################
cols = list(X.columns)
qr=pd.DataFrame({'ExplainedVariance':evr,'COLUMNS':cols})
qr
```

```python
qr = qr.sort_values('ExplainedVariance', ascending=False)
cols = list(qr.iloc[:,-1].head(9))
cols
print('\n THE MOST IMPORTANT COLUMNS OUT OF THE DATA AREA AS FOLLOWS:\n',cols)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(max_depth=2, random_state=0)  
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test) 
y_pred

# Performance Evaluation 

from sklearn.metrics import confusion_matrix  
from sklearn.metrics import accuracy_score

cm = confusion_matrix(y_test, y_pred)  
print(cm)  
print('Accuracy',accuracy_score(y_test, y_pred))  
DT1=round(100*(accuracy_score(y_test, y_pred)),2)
```

```python
#########################################     KERNEL SUPPORT VECTOR MACHINE     ##########################################

from sklearn.model_selection import train_test_split
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)
########################################################################################################################

# Predicting the test set results
from sklearn import metrics

Y_Pred = classifier.predict(X_test)

# Making the Confusion Matrix 

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, Y_Pred)
print("Accuracy of SVM:",metrics.accuracy_score(y_test, Y_Pred))
SVM=round(100*(metrics.accuracy_score(y_test, Y_Pred)),2)
print(cm)
```

```python
###############################################  NAIVE BAYES CLASSIFICATION    #######################################

from sklearn.model_selection import train_test_split
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,y_train)
########################################################################################################################

# Predicting the test set results
from sklearn import metrics

Y_Pred = classifier.predict(X_test)

# Making the Confusion Matrix 

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, Y_Pred)
print("Accuracy of Naive Bayes:",metrics.accuracy_score(y_test, Y_Pred))

NB=round(100*(metrics.accuracy_score(y_test, Y_Pred)),2)

print(cm)
```

```python
#######################################        K - N - N         #############################################################


from sklearn.model_selection import train_test_split
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)

########################################################################################################################

# Predicting the test set results
from sklearn import metrics

Y_Pred = classifier.predict(X_test)

# Making the Confusion Matrix 

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, Y_Pred)
print("Accuracy of KNN:",metrics.accuracy_score(y_test, Y_Pred))
print(cm)
KNN=round(100*(metrics.accuracy_score(y_test, Y_Pred)),2)
```

```python
print(DT1,SVM,NB,KNN)
lis=['DT1','SVM','NB','KNN']
score=DT1,SVM,NB,KNN
accuracyScores=pd.DataFrame({'MODELS':lis,'ACCURACY':score})
accuracyScores
```
